```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn 스타일 설정
sns.set(style='darkgrid')
font_path = "/Library/Fonts/Arial Unicode.ttf"  # 사용하는 운영체제에 따라 경로를 바꿔주세요.
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
# 데이터프레임에서 관련 열 선택
columns = ['A13_1', 'A13_2', 'A13_3', 'A13_4']

# 열 별로 교류 빈도를 카운트
count_data = df1[columns].apply(lambda x: x.value_counts()).T

# 효과적인 시각화를 위해 막대 그래프 사용
plt.figure(figsize=(14, 12))

# 각 막대에 다른 색상 설정
ax = sns.barplot(x=count_data.index, y=count_data[1], color='skyblue', label='전혀 아님')
ax = sns.barplot(x=count_data.index, y=count_data[2], color='lightgreen', bottom=count_data[1], label='1년에 1-2회')
ax = sns.barplot(x=count_data.index, y=count_data[3], color='lightcoral', bottom=count_data[1]+count_data[2], label='분기에 1-2회')
ax = sns.barplot(x=count_data.index, y=count_data[4], color='orange', bottom=count_data[1]+count_data[2]+count_data[3], label='월에 1-2회')
ax = sns.barplot(x=count_data.index, y=count_data[5], color='lightyellow', bottom=count_data[1]+count_data[2]+count_data[3]+count_data[4], label='주에 1-2회')
ax = sns.barplot(x=count_data.index, y=count_data[6], color='purple', bottom=count_data[1]+count_data[2]+count_data[3]+count_data[4]+count_data[5], label='거의 매일')

# x축의 이름 설정
plt.xlabel('물리적 교류 유형', fontsize=20)

# y축의 이름 설정
plt.ylabel('빈도', fontsize=20)

# 타이틀 추가
plt.title('다양한 그룹과의 물리적 교류 빈도 (A13)', fontsize=30)

# 범례 추가
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles=handles, title='빈도', title_fontsize='18', labels=['전혀 아님', '1년에 1-2회', '분기에 1-2회', '월에 1-2회', '주에 1-2회', '거의 매일'], loc='upper left',fontsize = '18')

# 플롯 보여주기
plt.show()
```
